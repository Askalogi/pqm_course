```python
import numpy
import matplotlib.pyplot as plt

# 1. Ορισμός της συναρτήσεως ορίου
def boundary_function(E, l):
    return numpy.cos(E)

# 2. Δημιουργία του φάσματος ενέργειας μέσα στο οποίο θα ελέγξουμε για μηδενικά
energy_range = numpy.linspace(-2.0*numpy.pi, 2.0*numpy.pi, 100)  # 100 σημεία μεταξύ 0 και 10
l = 1  # Η παράμετρος l μετατοπίζει την ημιτονοειδή καμπύλη πάνω/κάτω

# 3. Αρχικοποίηση της λίστας μηδενικών
zeros_list = []

# 4. Ξεκίνημα με την πρώτη τιμή
value = boundary_function(energy_range[0], l)

# 5. Επανάληψη σε κάθε ζεύγος διαδοχικών τιμών ενέργειας
for i in range(len(energy_range) - 1):
    new_value = boundary_function(energy_range[i + 1], l)

    # Έλεγχος για αλλαγή πρόσημου μεταξύ της τρέχουσας και της επόμενης τιμής
    if numpy.sign(new_value) * numpy.sign(value) == -1 and abs(value) < 1:
        # Αν αλλάζει το πρόσημο και η τιμή είναι "αρκετά κοντά" στο μηδέν, προσθέτουμε το διάστημα
        zeros_list.append((energy_range[i], energy_range[i + 1]))

    # Ενημέρωση της τιμής για τον επόμενο βρόχο
    value = new_value

# 6. Εκτύπωση των διαστημάτων στα οποία εντοπίστηκαν μηδενικά
print("Μηδενικά (αλλαγή πρόσημου της συνάρτησης) εντοπίστηκαν στα παρακάτω διαστήματα:")
for interval in zeros_list:
    print(f"  Μεταξύ E = {interval[0]:.4f} και E = {interval[1]:.4f}")
```

```python

# 7. Σχεδίαση γραφήματος
y_values = [boundary_function(E, l) for E in energy_range]

plt.figure(figsize=(10, 5))
plt.plot(energy_range, y_values, label='cos(E)', color='blue')
plt.axhline(0, color='gray', linestyle='--', label='y = 0')

plt.title("Μηδενικά της συνάρτησης cos(E)")
plt.xlabel("E (ακτίνια)")
plt.ylabel("cos(E)")
plt.legend()
plt.grid(True)
plt.show()
```
